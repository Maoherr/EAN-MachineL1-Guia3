## NIVELES DE OBESIDAD


La obesidad se clasifica en tres niveles según el Índice de Masa Corporal (IMC), que se calcula con la fórmula:

$$ \text{IMC} = \frac{\text{peso (kg)}}{\text{altura (m)}^2} $$

**Obesidad de nivel I (Obesidad moderada):**
- **IMC:** 30.0 - 34.9
- **Riesgos para la salud:** Aumenta el riesgo de enfermedades cardiovascuDares, diabeteII tipo 2, hipertensión y ciertos tipos de cáncer.

**Obesidad de nivel II (Obesidad severa):**
- **IMC:** 35.0 - 39.9
- **Riesgos para la salud:** Mayor riesgo de enfermedades cardiovDiabetes tipo IIetes tipo 2, hipertensión, apnea del sueño y problemas articulares.

**Obesidad de nivel III (Obesidad mórbida):**
- **IMC:** 40.0 o más
- **Riesgos para la salud:** Riesgo extremadamente alto de enfermedades carDiabetes tipo IIdiabetes tipo 2, hipertensión, apnea del sueño, problemas articulares graves y una reducción significativa de la esperanza de vida.

Es importante abordar la obesidad con un enfoque integral que incluya cambios en la dieta, aumento de la actividad física y, en algunos casos, intervenciones médicas o quirúrgicas. La prevención y el tratamiento temprano son clave para reducir los riesgos asociados con la obesidad.

## TIPOS DE DIABETES

**Diabetes Tipo I**  
La Diabetes Tipo I es una enfermedad autoinmune en la que el sistema inmunitario ataca y destruye las células beta del páncreas que producen insulina. Esto provoca que el cuerpo no pueda producir insulina, una hormona esencial para regular los niveles de glucosa en la sangre. Los síntomas incluyen sed excesiva, hambre constante, pérdida de peso inexplicable y fatiga. Aunque no tiene cura, se puede manejar con inyecciones de insulina, una dieta equilibrada y ejercicio regular.  

**Diabetes Tipo II**  
La Diabetes Tipo II es la forma más común de diabetes y se desarrolla cuando el cuerpo se vuelve resistente a la insulina o cuando el páncreas no produce suficiente insulina. Esto resulta en niveles elevados de glucosa en la sangre. Los factores de riesgo incluyen la obesidad, la falta de actividad física y antecedentes familiares. Los síntomas pueden ser similares a los de la Diabetes Tipo I, pero a menudo se desarrollan más lentamente. El tratamiento incluye cambios en el estilo de vida, medicamentos y, en algunos casos, insulina.

**Diabetes Tipo III**  
La Diabetes Tipo III no es una clasificación oficial, pero se utiliza para describir la relación entre la diabetes y la enfermedad de Alzheimer. Se cree que la resistencia a la insulina en el cerebro puede contribuir al desarrollo de la enfermedad de Alzheimer. Los síntomas incluyen problemas de memoria, dificultad para realizar tareas diarias y cambios en el comportamiento. Aunque no hay un tratamiento específico para la Diabetes Tipo III, se están investigando los efectos de los medicamentos antidiabéticos en esta condición.

## 4. Análisis de clustering y reducción de dimensionalidad

```python
# Importar las librerias necesarias
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Fijar semilla
np.random.seed(42)
```

```python
# Normalizar las variables numéricas
X = df[numeric_cols].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determinar número óptimo de clusters usando método del codo y silhouette
inertias = []
silhouette_scores = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, n_init ='auto', random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
# Graficar método del codo
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Método del Codo')

# Graficar silhouette scores
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'rx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs k')
plt.tight_layout()
plt.show()
```

El análisis de clusters usando K-means, determina visualmente que el número óptimo de clusters mediante el método del codo y el score de silhouette es **4**

## 5. Visualización en plano cartesiano usando PCA

```python
# Aplicar PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Determinar el número de clusters óptimo (según los gráficos anteriores)
optimal_clusters = 4
```

```python
# Entrenar el modelo KMeans con el número óptimo de clusters (k=4 basado en los gráficos)
kmeans = KMeans(n_clusters=optimal_clusters, n_init='auto', random_state=42)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
```

```python
# 5. Visualizaciones y reducciones de dimensionalidad
pca = PCA(n_components=2)
df_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(13, 6))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=clusters, palette='viridis')
plt.title('PCA de las Variables Numéricas')
plt.xlabel('Primera Componente Principal')
plt.ylabel('Segunda Componente Principal')
plt.show()
```

```python
# Mostrar varianza explicada por componentes principales
print("\nVarianza explicada por componentes principales: ", pca.explained_variance_ratio_ )
```

```python
# Correlación entre clusters y obesidad
print("\nDistribución de clusters por clase de obesidad:")
print(pd.crosstab(clusters, df['NObeyesdad']))
```
